```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("darkgrid")
```

```python
df = pd.read_csv("timing_sweep_f3c6afc3-018e-eb11-a2ce-c4346bcb1550.csv")
```

```python
timings_df = pd.read_csv("../southern_term_sweep_scenarios.csv")
df = pd.merge(df,timings_df, on="scenario_number")
```

```python
df
```

```python
t1 = df.groupby("term1_day").agg({"pfpr2_10": "mean"}).reset_index()
t2 = df.groupby("term2_day").agg({"pfpr2_10": "mean"}).reset_index()
t3 = df.groupby("term3_day").agg({"pfpr2_10": "mean"}).reset_index()
```

```python
baseline_pfpr = np.mean([
    0.2733096412561,
    0.2692578415019,
    0.2682835010497
])
```

```python
plt.figure(dpi=200)
plt.plot(t1["term1_day"], t1["pfpr2_10"], marker='o', label="Term 1")
plt.plot(t2["term2_day"], t2["pfpr2_10"], marker='o', label="Term 2")
plt.plot(t3["term3_day"], t3["pfpr2_10"], marker='o', label="Term 3")
plt.axhline(baseline_pfpr, label="Baseline (no IPT)", color="C4", linestyle='dashed')
plt.xlabel("Day")
plt.ylabel("PfPR2-10")
plt.legend()
plt.ylim([0.2,0.28])
plt.savefig("southern_timing_sweep_example_v2.pdf")
```

```python
df_refine = df[np.logical_and(df["term2_day"]==165,df["term3_day"]==254)]
t1_refine = df_refine.groupby("term1_day").agg({"pfpr2_10": "mean"}).reset_index()

plt.plot(t1["term1_day"], t1["pfpr2_10"], marker='o', label="Term 1")
plt.plot(t1_refine["term1_day"], t1_refine["pfpr2_10"], marker='o', label="Term 1 refined")
plt.legend()
```

```python
best_timing = df[np.logical_and.reduce([
    df["term1_day"]==35,
    df["term2_day"]==165,
    df["term3_day"]==254])]
```

```python
best_timing
```

```python
plt.figure(dpi=200)
sns.boxplot(df["pfpr2_10"])
plt.axvline(np.median(best_timing["pfpr2_10"]), label="Best timing", color="C3")
plt.axvline(baseline_pfpr,label="Baseline (no IPT)", color="C4")
plt.legend()
```
