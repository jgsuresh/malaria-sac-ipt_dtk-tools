```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("white")
```

```python
df_cent = pd.read_csv("timeseries_53be112d-06b8-eb11-a2e3-c4346bcb7275.csv")
df_sahel = pd.read_csv("timeseries_312b5201-609d-eb11-a2ce-c4346bcb1550.csv")
df_south = pd.read_csv("timeseries_f70751f3-2d98-eb11-a2ce-c4346bcb1550.csv")
df = pd.concat([df_cent, df_sahel, df_south], ignore_index=True)
```

```python
foo = df.groupby(["scenario_number", "burnin_approx_pfpr2_10", "archetype", "time"]).agg({"prev": "mean", "cases": "median"}).reset_index()
```

```python
a = "Central"
p = 0.2
plt.figure(dpi=300)
fooa = foo[foo["archetype"]==a]
foob = fooa[fooa["burnin_approx_pfpr2_10"]==p]
for s, sdf in foob.groupby("scenario_number"):
    if s == 13:
        label="Baseline + IPTsc"
    if s == 30:
        label="Baseline interventions"
    plt.plot(sdf["time"], sdf["cases"], label=label)
for y in [0,1]:
    for d in [60,264,349]:
        plt.axvline(y*365+d,linestyle='dashed',color='purple', alpha=0.3)
plt.legend()
plt.xlabel("Day")
plt.ylabel("Cases per day")
plt.title("{} - burnin pfpr2-10 = {}".format(a,p))
plt.ylim([0,40])
```
