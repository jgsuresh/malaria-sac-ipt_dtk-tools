```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
sns.set_style("darkgrid")
```

```python
df = pd.read_csv("zambia_gridded_itn_events.csv")
lookup = pd.read_csv("grid_lookup.csv")

df = df.merge(lookup[["grid_cell", "pop"]], on="grid_cell", how="left")
```

```python
df["year"] = df["fulldate"].apply(lambda x: np.float(x[:4])).astype(np.int)
```

```python
df["pop_cov"] = df["cov_all"]*df["pop"]
```

```python
foo = df.groupby("year").agg({"pop": "sum", "pop_cov": "sum"}).reset_index()
foo["cov_all"] = foo["pop_cov"]/foo["pop"]
```

```python
foo
```

```python
plt.figure(dpi=200)
plt.scatter(foo["year"], foo["cov_all"])
plt.ylim([0,1])
```

```python
amelia_df = pd.read_csv("C:/Users/jsuresh/OneDrive - Institute for Disease Modeling/Projects/SAC IPT/ITN coverages/fig_2_access_use_timeseries.csv")
amelia_df = amelia_df[amelia_df["country_name"]=="Zambia"]
amelia_use = amelia_df[amelia_df["variable"]=="use"].reset_index(drop=True)
amelia_access = amelia_df[amelia_df["variable"]=="access"].reset_index(drop=True)
```

```python
plt.figure(dpi=200)
plt.plot(foo["year"], foo["cov_all"], label="Kariba coverage from grid model", marker='o')

plt.plot(amelia_use["time"], amelia_use["mean_among_atrisk"], label="use", linestyle='dashed')
plt.plot(amelia_access["time"], amelia_access["mean_among_atrisk"], label="access", linestyle='dashed')
plt.plot(amelia_use["time"], amelia_use["mean_among_atrisk"]*1/0.869/0.75, label="inferred coverage")
plt.ylim([0,1])
plt.legend()
```

```python
amelia_use_binned = amelia_use.groupby("year").agg({"mean_among_atrisk": "mean",
                                                   "lower_among_atrisk": "mean",
                                                   "upper_among_atrisk": "mean"}).reset_index()
amelia_use_binned["inferred_coverage"] = amelia_use_binned["mean_among_atrisk"]*1/0.869/0.75
amelia_use_binned["inferred_coverage_low"] = amelia_use_binned["lower_among_atrisk"]*1/0.869/0.75
amelia_use_binned["inferred_coverage_high"] = amelia_use_binned["upper_among_atrisk"]*1/0.869/0.75
```

```python
plt.figure(dpi=200)
plt.plot(foo["year"], foo["cov_all"], label="Kariba coverage from survey", marker='o')

# plt.plot(amelia_use["time"], amelia_use["mean_among_atrisk"], label="use", linestyle='dashed')
# plt.plot(amelia_use_binned["year"], amelia_use_binned["mean_among_atrisk"], label="use, binned", linestyle='dashed', marker='o')
plt.plot(amelia_use_binned["year"], amelia_use_binned["inferred_coverage"], label="Coverage inferred from Amelia usage", marker='o')
plt.fill_between(amelia_use_binned["year"], 
                 amelia_use_binned["inferred_coverage_low"],
                 amelia_use_binned["inferred_coverage_high"],
                color="C1",alpha=0.4)

plt.ylim([0,1])
plt.legend(fontsize=10)
```

## Trying to create a single historical ITN usage file

```python
comparison_df = pd.merge(foo, amelia_use_binned, on="year", how="inner")
```

```python
comparison_df["difference_factor"] = comparison_df["cov_all"]/comparison_df["inferred_coverage"]
```

```python
plt.plot(comparison_df["difference_factor"])
plt.axhline(np.mean(comparison_df["difference_factor"]))
print(np.mean(comparison_df["difference_factor"]))
correction_factor = np.mean(comparison_df["difference_factor"])
```

```python
plt.figure(dpi=200)
plt.plot(foo["year"], foo["cov_all"], label="Kariba coverage from survey", marker='o')

# plt.plot(amelia_use["time"], amelia_use["mean_among_atrisk"], label="use", linestyle='dashed')
# plt.plot(amelia_use_binned["year"], amelia_use_binned["mean_among_atrisk"], label="use, binned", linestyle='dashed', marker='o')
cut = np.logical_or(amelia_use_binned["year"] < 2009, amelia_use_binned["year"] > 2017)
plt.scatter(amelia_use_binned[cut]["year"], amelia_use_binned[cut]["inferred_coverage"]*correction_factor, label="Coverage inferred from Amelia usage", marker='o',c="C1")


plt.ylim([0,1])
plt.legend(fontsize=10)
```

```python
spline1 = amelia_use_binned[cut].reset_index(drop=True)
spline1["cov_all"] = spline1["inferred_coverage"]*correction_factor
spline1 = spline1[["year", "cov_all"]]
spline1["source"] = "amelia usage with correction"

spline2 = foo[["year", "cov_all"]]
spline2["source"] = "kariba survey"
# spline_df = pd.concat([])

spline_df = pd.concat([spline1, spline2]).sort_values(by="year")
```

```python
plt.plot(spline_df["year"], spline_df["cov_all"])
```

```python
spline_df.to_csv("southern_historical_itn.csv", index=False)
```
